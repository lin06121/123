## 数据集迭代

数据集加载后，一般以迭代方式获取数据，送入神经网络中进行训练。

我们可以用`create_tuple_iterator`或`create_dict_iterator`接口创建数据迭代器，迭代访问数据。

- create_tuple_iterator性能略高，
- create_dict_iterator使用键访问更直观

访问的数据类型默认为`Tensor`；若设置`output_numpy=True`，访问的数据类型为`Numpy`。

下面定义一个可视化函数，迭代9张图片进行展示。

```python
def visualize(dataset):
    figure = plt.figure(figsize=(4, 4))
    cols, rows = 3, 3

    plt.subplots_adjust(wspace=0.5, hspace=0.5)
    #设置子图都水平/垂直间距，值0.5表示子图宽度/高度的50%将用作子图之间的空间。
    for idx, (image, label) in enumerate(dataset.create_tuple_iterator()):
        figure.add_subplot(rows, cols, idx + 1)
        plt.title(int(label))
        plt.axis("off")  #不显示刻度线和边框
        plt.imshow(image.asnumpy().squeeze(), cmap="gray")  #squeeze()：去除数组中维度为 1 的轴（例如将形状从 (1, 28, 28) 压缩为 (28, 28)）
        if idx == cols * rows - 1:
            break
    plt.show()
```

```python
visualize(train_dataset)
```

## 数据集常用操作

数据集常采用`dataset = dataset.operation()`的异步执行方式，此时不执行操作，在Pipeline中加入节点，最终进行迭代时，行执行整个Pipeline。

Pipeline的节点包括但不限于：
1. shuffle：对数据进行随机打乱操作，比如对图片进行随机打乱等操作；
2. map：对数据进行映射操作，比如对图片进行resize、crop等操作；
3. batch：对数据进行批量操作，比如对图片进行批量化等操作； 

下面分别介绍几种常见的数据集操作。

### shuffle

![op-shuffle](https://mindspore-website.obs.cn-north-4.myhuaweicloud.com/website-images/r2.3/tutorials/source_zh_cn/advanced/dataset/images/op_shuffle.png)

数据集随机`shuffle`可以消除数据排列造成的分布不均问题。

`mindspore.dataset`提供的数据集在加载时可配置`shuffle=True`，或使用如下操作：

```python
train_dataset = train_dataset.shuffle(buffer_size=64)  #buffer_size越大，打乱的随机性越强，更多的样本被混合在一起。

visualize(train_dataset)
```

### map

`map`操作是数据预处理的关键操作，可以针对数据集添加数据变换，将数据变换应用于该列数据的每个元素，并返回包含变换后元素的新数据集。 

这里我们对Mnist数据集做数据缩放处理，将图像统一除以255，数据类型由uint8转为了float32。

```python
train_dataset = train_dataset.map(vision.Rescale(1.0 / 255.0, 0), input_columns='image')
```

对比map前后的数据，可以看到数据类型变化。

```python
image, label = next(train_dataset.create_tuple_iterator())
print(image.shape, image.dtype)
```

### batch

将数据集打包为固定大小的`batch`是在有限硬件资源下使用梯度下降进行模型优化的折中方法，可以保证梯度下降的随机性和优化计算量。

![op-batch](https://mindspore-website.obs.cn-north-4.myhuaweicloud.com/website-images/r2.3/tutorials/source_zh_cn/advanced/dataset/images/op_batch.png)

一般我们会设置一个固定的batch size，将连续的数据分为若干批（batch）。

```python
train_dataset = train_dataset.batch(batch_size=32)
```

batch后的数据增加一维，大小为`batch_size`。

```python
image, label = next(train_dataset.create_tuple_iterator())
print(image.shape, image.dtype)
```

## 自定义数据集

`mindspore.dataset`模块提供了一些常用的公开数据集和标准格式数据集的加载API。

对于MindSpore暂不支持直接加载的数据集，可以构造自定义数据加载类或自定义数据集生成函数的方式来生成数据集，然后通过`GeneratorDataset`接口实现自定义方式的数据集加载。

`GeneratorDataset`支持通过可随机访问数据集对象、可迭代数据集对象和生成器构造自定义数据集，下面分别对其进行介绍。

### 可随机访问数据集

可随机访问数据集是实现了`__getitem__`和`__len__`方法的数据集，表示可以通过索引/键直接访问对应位置的数据样本。

```python
# Random-accessible object as input source
class RandomAccessDataset:
    def __init__(self):
        self._data = np.ones((5, 2))
        self._label = np.zeros((5, 1))

    def __getitem__(self, index):
        return self._data[index], self._label[index]

    def __len__(self):
        return len(self._data)
```

```python
loader = RandomAccessDataset()
dataset = GeneratorDataset(source=loader, column_names=["data", "label"])
#GeneratorDataset会自动调用loader的__getitem__和__len__方法
#针对每个索引，获取相应的数据和标签，并分别存储在名为"data"和"label"的列中

for data in dataset:
    print(data)
```

当使用`dataset[idx]`访问这样的数据集时，可以读取dataset内容中第idx个样本或标签。

```python
class RandomAccessDataset:
    def __init__(self):
        self._data = np.array([1,2,3])
        self._label = np.array([0,1,0])

    def __getitem__(self, index):
        return self._data[index], self._label[index]

    def __len__(self):
        return len(self._data)

loader = RandomAccessDataset()   
  
data_item,label_item = loader[2]
print(data_item,label_item)

#dict_iterator = dataset.create_dict_iterator(output_numpy=True) 创建字典迭代器
#all_data = list(dict_iterator)
#print(all_data)
```

也支持列表、元组、numpy数据类型

```python
loader = [np.array(0), np.array(1), np.array(2)]
dataset = GeneratorDataset(source=loader, column_names=["data"])

for data in dataset:
    print(data)

# GeneratorDataset构造函数默认设置shuffle=True
# 这会导致数据集中的样本在每次迭代时被随机打乱
```

### 可迭代数据集

可迭代的数据集是实现了`__iter__`和`__next__`方法的数据集，表示可以通过迭代的方式逐步获取数据样本。适合流式数据、大型数据集或无法预先知道总长度的数据源。只能按顺序访问，无法直接跳转到特定位置。

例如，当使用`iter(dataset)`的形式访问数据集时，可以读取从数据库、远程服务器返回的数据流。

下面构造一个简单迭代器，并将其加载至`GeneratorDataset`。

```python
# Iterator as input source

class IterableDataset():
    def __init__(self, start, end):
        '''init the class object to hold the data'''
        self.start = start
        self.end = end
    def __next__(self):
        '''iter one data and return'''
        return next(self.data)
    def __iter__(self):
        '''reset the iter'''
        self.data = iter(range(self.start, self.end))
        return self
```

```python
loader = IterableDataset(1, 5)
dataset = GeneratorDataset(source=loader, column_names=["data"])

#当迭代dataset时，MindSpore会调用loader的__iter__方法开始迭代
#每次需要新数据时，调用loader的__next__方法获取下一个值

for d in dataset:
    print(d)
```

用可迭代数据集创建data和label组合数据集

```python
class IterableDatasetWithLabels:
    def __init__(self, start, end):
        self.start = start
        self.end = end
        
    def __iter__(self):
        self.index = self.start
        return self
        
    def __next__(self):
        if self.index >= self.end:
            raise StopIteration
        
        # 创建数据和标签对
        data = np.array([self.index, self.index * 2])  # 示例特征
        label = np.array([self.index % 2])  # 示例标签
        
        self.index += 1
        return data, label  # 返回(data, label)元组

loader = IterableDatasetWithLabels(1, 5)
dataset = GeneratorDataset(source=loader, column_names=["data", "label"])

for item in dataset:
    print(f"数据: {item[0]}, 标签: {item[1]}")
```

```python
#也支持多列数据，例如，特征、标签和权重三个列
class MultiColumnIterator:
    def __init__(self, start, end):
        self.start = start
        self.end = end
        
    def __iter__(self):
        self.index = self.start
        return self
        
    def __next__(self):
        if self.index >= self.end:
            raise StopIteration
            
        # 创建三个数据列
        feature = np.array([self.index, self.index * 2])
        label = np.array([self.index % 2])
        weight = np.array([1.0 if self.index % 2 == 0 else 0.5])
        
        self.index += 1
        return feature, label, weight

# 创建三列数据集
loader = MultiColumnIterator(1, 5)
dataset = GeneratorDataset(
    source=loader, 
    column_names=["feature", "label", "weight"]
)

for item in dataset:
    print(f"特征: {item[0]}, 标签: {item[1]}, 权重: {item[2]}")
```

### 生成器

也可以使用Python的生成器函数实现同样的功能。
`source`参数使用 `lambda`

```python
def data_and_label_generator(start, end):
    for i in range(start, end):
        data = np.array([i, i * 2])  # 示例特征数据
        label = np.array([i % 2])    # 示例标签数据
        yield data, label  # 同时生成数据和标签

# 使用lambda函数包装生成器(因为生成器只能迭代一次)
dataset = GeneratorDataset(
    source=lambda: data_and_label_generator(1, 5),
    column_names=["data", "label"]
)

for item in dataset:
    print(f"数据: {item[0]}, 标签: {item[1]}")
```
